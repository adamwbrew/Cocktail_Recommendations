```python
def NormalizeInstructions(steps, delimiter, prep = None):
    """
    steps = string of steps for making drink, Recipe
    delimiiter = offset character between steps
    prep = Extra optional step required to create drink

    returns Recipe = list of strings in order of recipe for drink
    """
    Recipe = []
    has_period = steps.find(".")
    if(has_period == -1):
        steps = steps + "."
        steps_list = steps.split(f"{delimiter}")
    else:
        steps_list = steps.split(f"{delimiter}")
    steps_list = [f"{i+1}: {steps_list[i].strip()}." for i in range(len(steps_list))]
    steps_list.pop()
    # prep = str(prep).strip()
    if(prep != None and prep != ''):
        prep = "Prep: " + str(prep).strip()
        Recipe = [prep] + steps_list
    else:
        Recipe = steps_list
    return Recipe
```

```python
def LiquorType(ingredient_list):   
    alcohol_types = ["brandy", "gin", "rum", "rhum","tequila", "vodka", "whiskey", "whisky", "bourbon", "wine", "champagne",
     "cognac", "absinthe", "scotch", "pisco", "rye", "king's ginger liqueur", "mezcal", "luxardo maraschino liqueur",
     "luxardo aperitivo", "h by hine", "ypioca brasilizar cachaca prata classica", "ypioca brasilizar cachaca prata reserva especial", 
     "glenrothes vintage reserve", "prosecco", "jack daniel's", "the king's ginger", "christmas spirit"]
    alcohol = ''
    for item in ingredient_list:
        if(alcohol != ''):
            break
        item_listed = item.split(" ")
        for element in item_listed:
            element = element.replace(",", '')
            element = element.replace(".", '')
            element = element.replace(" ", '')
            element = element.replace("#", 'No.')
            if(element.lower() in alcohol_types):
                alcohol = element.title()
    if(alcohol == ''):
        for item in ingredient_list:
            item_listed = item.split(" ")
            spacers = len(item_listed)
            for i in range(spacers):
                item_check = item_listed[i:spacers]
                item_broken = (' '.join(item_check)).lower()
                if(item_broken in alcohol_types):
                    alcohol = item_broken.title()
    if(alcohol == ''):
        alcohol = "Other"
    return alcohol
```

```python
def GlasswareNormalize(glass):
    glass = str(glass).strip()
    if(glass == "" or glass == "nan"):
        glass = "Unknown"
    else:
        glass_broken = glass.split(" ")
        end = glass_broken[-1].lower()
        if(end != "glass" and end != "mug"):
            glass_broken.append("glass")
        glass = ' '.join(glass_broken)
    return glass
```

```python
def measurementsMaker(Ingredient):
    if(Ingredient == None or Ingredient == 'nan' and Ingredient == np.nan):
        measure = None
    else:
        measure = ''
        measurement_types = ["oz", "oz.", "ml", "l", "gram", "lbs", "lb", "dash", "pinch", "dashes", "splash", "scoop", "cup", "cups", "part", "cube"]
        measures_dict = {".125":"0.125", ".16":"0.16",".33":"0.33",".34":"0.33",".375":"0.375",".625":"0.625", "¼":"0.25", "1/4":"0.25", ".25":"0.25","1/3":"0.33", "1/2":"0.5", ".5":"0.5", "½":"0.5", "2/3":"0.67", ".67":"0.67", "3/4":"0.75", ".75":"0.75", "¾":"0.75", "1½":"1.5"}
        measurement_finder = Ingredient.strip().lower().split(" ")
        for j in range(len(measurement_finder)):
            if(measurement_finder[j] in measures_dict.keys()):
                measurement_finder[j] = measures_dict[measurement_finder[j]]
        count = 0
        for j in measurement_finder:
            if(j.isalpha()):
                count += 1
        if(count == len(measurement_finder)):
            if(measurement_finder[0] == "several" and measurement_finder[1] in measurement_types):
                measure = measurement_finder[0] + ' ' + measurement_finder[1]
            elif(measurement_finder[0] == "several"):
                measure = measurement_finder[0]
            else:
                measure = None
        else:
            for i in range(len(measurement_finder)):
                word = measurement_finder[i]
                # find measurement type
                if(word in measurement_types):
                    # check if measure amount id in two parts (Ex: 1 3/4 oz)
                    if(measurement_finder[0].replace(".", "").isdigit() and measurement_finder[1].replace(".", "").isdigit()):
                        numeric = str(float(measurement_finder[0]) + float(measurement_finder[1]))
                        measure = (numeric + ' ' + measurement_finder[i]).lower()
                    elif(measurement_finder[i-1] not in measurement_types and measurement_finder[i-1].isalpha()):
                        measure = measurement_finder[0].lower()
                    else:
                        measure = ' '.join(measurement_finder[:i+1]).lower()
                    break
            if(measurement_finder[0].isdigit() and (measurement_finder[1] not in measurement_types) and measure == ""):
                measure = measurement_finder[0]
            if(measure == ''):
                measure = None
        if(measure != None):
            oz_changer = measure.lower().split(' ')
            if("oz." in oz_changer):
                oz_changer[-1] = 'oz'
                measure = ' '.join(oz_changer).lower()

    return measure
```

```python
def measurementsRemover(Ingredient):
    
    if(Ingredient == None or Ingredient == 'nan' and Ingredient == np.nan):
        result = None
    else:
        result = ''
        measurement_types = ["oz", "oz.", "ml", "l", "gram", "lbs", "lb", "dash", "pinch", "dashes", "splash", "scoop", "cup", "cups", "part", "cube", "1", "2"]
        measurement_finder = Ingredient.lower().split(" ")
        if(len(measurement_finder) == 1):
            result = (measurement_finder[0]).title()
        else: 
            for i in range(len(measurement_finder)):
                word = measurement_finder[i]
                if(word in measurement_types):
                    if(measurement_finder[i-1] != "several" and measurement_finder[i-1].isalpha()):
                        # print(word)
                        result = ' '.join(measurement_finder[1:]).title()
                    else:
                        result = ' '.join(measurement_finder[i+1:]).title()

    return result
```
