```python
import requests
from API_secrets import API_Key
import pandas as pd
```

```python
cocktails = []
letters = 'abcdefghijklmnopqrstuvwxyz'
for letter in letters:
    api_url = f'https://www.thecocktaildb.com/api/json/v1/1/search.php?f={letter}'
    response = requests.get(api_url, headers={'X-Api-Key': f'{API_Key}'})
    if response.status_code == requests.codes.ok:
        drinks = response.json()["drinks"]
    else:
        print("Error:", response.status_code, response.text)

    if(drinks == None): continue
    else:
        cocktails_mini = []
        for drink in drinks:
            if(drink["strAlcoholic"] == 'Alcoholic'):
                cocktail = drink
                del cocktail["idDrink"]
                del cocktail["strCategory"]
                del cocktail["strDrinkAlternate"]
                del cocktail["strTags"]
                del cocktail["strIBA"]
                del cocktail["strAlcoholic"]
                del cocktail["strInstructionsES"]
                del cocktail["strInstructionsDE"]
                del cocktail["strInstructionsFR"]
                del cocktail["strInstructionsIT"]
                del cocktail["strInstructionsZH-HANS"]
                del cocktail["strInstructionsZH-HANT"]
                del cocktail["strDrinkThumb"]
                del cocktail["strImageSource"]
                del cocktail["strImageAttribution"]
                del cocktail["strCreativeCommonsConfirmed"]
                del cocktail["dateModified"]
                del cocktail["strVideo"]
                cocktails_mini.append(cocktail)
        cocktails += cocktails_mini
print(len(cocktails))
```

```python
def NormalizeInstructions(steps, delimiter, prep = None):
    """
    steps = string of steps for making drink, Recipe
    delimiiter = offset character between steps
    prep = Extra optional step required to create drink

    returns Recipe = list of strings in order of recipe for drink
    """
    Recipe = []
    has_period = steps.find(".")
    if(has_period == -1):
        steps = steps + "."
        steps_list = steps.split(f"{delimiter}")
    else:
        steps_list = steps.split(f"{delimiter}")
    steps_list = [f"{i+1}: {steps_list[i].strip()}." for i in range(len(steps_list))]
    steps_list.pop()
    # prep = str(prep).strip()
    if(prep != None and prep != ''):
        prep = "Prep: " + str(prep).strip()
        Recipe = [prep] + steps_list
    else:
        Recipe = steps_list
    return Recipe
```

```python
# normalize all instructions
cocktails_df["Instructions"] = cocktails_df.apply(lambda x : NormalizeInstructions(x["Instructions"], ".",), axis=1)
# remove used columns and uneeded columns
cocktails_df.head()
```

```python
def LiquorType(ingredient_list):   
    alcohol_types = ["Beer", "brandy", "gin", "rum", "rhum","tequila", "vodka", "whiskey", "whisky", "bourbon", "wine", "champagne",
     "cognac", "absinthe", "scotch", "pisco", "rye", "king's ginger liqueur", "mezcal", "luxardo maraschino liqueur",
     "luxardo aperitivo", "h by hine", "ypioca brasilizar cachaca prata classica", "ypioca brasilizar cachaca prata reserva especial", 
     "glenrothes vintage reserve", "prosecco", "jack daniel's", "the king's ginger", "christmas spirit", "schnapps", "Baileys irish cream",
     "Triple sec", "Vermouth", "Jack Daniels", "Everclear", "liqueur", "Wild Turkey", "Cointreau"]
    alcohol_types = [x.lower() for x in alcohol_types]
    alcohol = ''
    for item in ingredient_list:
        if(alcohol != ''):
            break
        item_listed = item.split(" ")
        for element in item_listed:
            element = element.replace(",", '')
            element = element.replace(".", '')
            element = element.replace(" ", '')
            element = element.replace("#", 'No.')
            if(element.lower() in alcohol_types):
                alcohol = element.title()
    if(alcohol == ''):
        for item in ingredient_list:
            item_listed = item.split(" ")
            spacers = len(item_listed)
            for i in range(spacers):
                item_check = item_listed[i:spacers]
                item_broken = (' '.join(item_check)).lower()
                if(item_broken in alcohol_types):
                    alcohol = item_broken.title()
    if(alcohol == ''):
        alcohol = "Other"
    return alcohol
```
